```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
#### TEMPS D'EXECUTIONS ###
### graphe avec 500 sommets
edges = [1000,2000,5000,10000,50000,100000,120000]
d = [.008,.016,.0401,.0802,.4008,.8016,.9619]
times = [0,0,0,0,4,29,16.2]
fig, (ax1, ax2) = plt.subplots(1,2,figsize=(13,4))
ax1.set(xlabel='|E|', ylabel='exec time (s)', label='time/|E|')
ax2.set(xlabel='density', ylabel='exec time (s)',label='time/density')

ax1.plot(edges,times)
ax2.plot(d,times)

fig.show()
```
